```python
import pandas as pd
import requests
```

Using requests the function will call the http link thorugh api which will give our desire data

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
```

converting into JSON format

```python
response.json()
```

For specific segment use results

```python
response.json()['results']
```

giving it a frame row/column

```python
pd.DataFrame(response.json()['results'])
```

working with some portion use head(number of index want to display)

```python
pd.DataFrame(response.json()['results']).head(2)
```

for displaying required field

```python
pd.DataFrame(response.json()['results']).head(2)[['id','title','release_date']]
```

whole thing in a single variable

```python
df = pd.DataFrame(response.json()['results']).head(2)[['id','title','release_date']]
df.head()
```

for survey each page

```python
for i in range(1,429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = pd.concat([df, temp_df])
    #df = df.append(temp_df,ignore_index=True)
  #igonre will maintain indexing
```

```python
df
```

```python
for i in range(1,429):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = pd.concat([df, temp_df], ignore_index=True)
  #igonre will maintain indexing
```
