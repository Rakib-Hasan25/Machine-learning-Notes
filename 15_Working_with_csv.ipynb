3. Opening a CSV file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {'User-Agent': "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}

req = requests.get(url, headers= headers)

print(type(req))
print(type(req.text))  #string
data = StringIO(req.text) #StringIO() makes string to io.StringIO ,
#from this formet of data we can read as dataframe
print(type(data))
df = pd.read_csv(data)
df.head()
```

9. Skiprows/nrows paramter

```python
# pd.read_csv('stock_data.csv', skiprows=[1,3,5]) #deletes 1st, 3rd and 5th rows
# pd.read_csv('stock_data.csv', skiprows=range(1, 3))
# pd.read_csv('stock_data.csv', nrows=3)
def custom(row_ind):
    return row_ind % 2 == 1
pd.read_csv('stock_data.csv', skiprows=custom) # using custom function in skiprows
```
