```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
df = pd.read_csv('train.csv')[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.sample(10)
```

```python
df.dropna(inplace=True)



# drop all null value

#  when inplace=True is set, the DataFrame is modified in place, meaning
#  that the changes are made directly to the original DataFrame object,
#  and the operation does not return a new DataFrame.
```

```python
df.sample(5)
```

```python
# separate input and output column and take all the row

X = df.iloc[:,0:4]
y = df.iloc[:,-1]



# [:,0:4] means selecting all rows (: before the comma)
# and the columns from index 0 up to (but not including) index 4 (0:4 after the comma)
```

```python
X.head()
```

```python
cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20)


# cross_val_score: This function performs cross-validation to evaluate the performance of a model.
# It takes several arguments:
# The first argument (LogisticRegression()): This is the model you want to evaluate.
# The second argument (X): This represents the input features of your dataset.
# The third argument (y): This represents the target variable you're trying to predict.



# scoring='accuracy'--->we're telling the computer to
# measure how often the model's predictions are correct compared to the actual outcomes.
# example:
#  Imagine you have a test with 100 questions,
#   and you answer 85 of them correctly. Your accuracy
#   is the proportion of correct answers, which is 85/100 = 0.85.





#  cv=20 means we're doing something similar. We're splitting our data into 20 parts or group,
#   training our model on 19 group of them, and testing it on the remaining one group. We do this 20 times,
#   rotating which part we use for testing each time. This helps us get a good overall idea of how well our model works.
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

#Applying Feature Contruction

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1

# +1 for the person itself
```

```python
X.head()
```

```python
def myfunc(num):
    if num == 1:
        #alone
        return 0
    elif num >1 and num <=4:
        # small family
        return 1
    else:
        # large family
        return 2
```

```python
X['Family_type'] = X['Family_size'].apply(myfunc)
```

```python
X.sample(5)
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

#Feature Splitting

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df['Name']
```

```python
# df['Name'].str.split(', ', expand=True)
```

```python
# df['Name'].str.split(', ', expand=True)[1]
```

```python
df['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
df['Title']
```

```python
df.info()
```

```python
# grouping by 'Title', this part selects the 'Survived' column from each group.
# : This calculates the mean (average) of the values in the 'Survived' column within each group.

new=df.groupby('Title')['Survived'].mean()
new
```

```python
# (df.groupby('Title').mean()['Survived']).sort_values(ascending=False) not working

df.groupby('Title')['Survived'].mean().sort_values(ascending=False)


# here we can see :
# Mrs -> survival rate is 0.792
# Miss-> survival rate is 0.69

# we create an extra column which is useful in feature engineering
```

```python
df['Is_Married'] = 0
df['Is_Married'].loc[df['Title'] == 'Mrs'] = 1
df
```
